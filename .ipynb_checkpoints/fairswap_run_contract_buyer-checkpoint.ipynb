```python
from web3 import Web3

# 連接ganache
ganache_url = "http://localhost:7545"
w3 = Web3(Web3.HTTPProvider(ganache_url))

print(f'連接狀態: {w3.is_connected()}')
```

```python
import json

# 加載智能合約ABI
with open('abi7.json', 'r') as file:
    contract_abi = json.load(file)

# 創建智能合約實例
with open("contract_address.txt", "r") as f:
    contract_address = f.read()

contract = w3.eth.contract(
    address=contract_address,
    abi=contract_abi
)
print(contract)

# 設置參數(lab)
sender_address = '0x7439a45be3a121e74Ef79b504dE0200b8260D9c3'
#sender_private_key = '0x6f66f3fa08379a98c0e1f92e6a9dedf1252a503f4faf33c45b51e4bdc121840a'

receiver_address = '0xc7123a13098d296293095e9956479766F634fCCE'
receiver_private_key = '0xe0200986d8689b14eb92e5dba6847abce0254bc39866ec06d8a986a2ea97db5d'

# 設置參數(home)
#sender_address = '0xD2ba6B552Aa333b15b69DB2CB173C8c50415e0cd'
#sender_private_key = '0x4e768f45a9106f44ee9718de6eaee54ec09b0dc72a835f9a17c6848a0db55955'

#receiver_address = '0x2Dd7bF317cbeb7Fe7C6d837C68AbA01B0EE60D21'
#receiver_private_key = '0x321d0f0b6bbd53f5d8c42e85f462a37add9677cfb100d3f7ab8571f6b28ff659'

price = 1
```

# def

```python
# 查詢錢包餘額
def balance(account):
    balance_wei = w3.eth.get_balance(account)
    balance = balance_wei / (10 ** 18)
    print(account,'的錢包餘額:', balance)
    
# 獲取stage
def get_stage():
    current_stage = contract.functions.phase().call()
 
    stage_names = ['created', 'initialized', 'accepted', 'keyRevealed', 'finished']
    current_stage_name = stage_names[current_stage]

    print("當前的stage:", current_stage_name)
    
# 獲取智能合約餘額
def get_smartcontract_balance():
    contract_balance = w3.eth.get_balance(contract_address)
    
    contract_balance_eth = w3.from_wei(contract_balance, 'ether')

    print(f"合约的餘額是 {contract_balance_eth} ETH")

balance(receiver_address)
get_stage()
get_smartcontract_balance()
```

# 執行合約


```python
with open("rz.txt", "rb") as f:
    rz = f.read()
ciphertextRoot = w3.to_hex(rz)
#fileRoot = '0xfa7e5f3d12a99e04e14828cd50e9e1ca0cc26056ae442463252c0bf3e0992ce2'

print('ciphertextRoot:', ciphertextRoot)
```

```python
# Step 2 接受交易

# 獲取交易的nonce值
nonce = w3.eth.get_transaction_count(receiver_address)

# 構建交易參數
transaction_params = {
    'gas': 500000,
    'gasPrice': w3.to_wei('10', 'gwei'),
    'nonce': nonce,
    'from': receiver_address,
    'value': w3.to_wei('1', 'ether')
}
try:
    # 構建交易數據
    transaction_data = contract.functions.accept(ciphertextRoot).build_transaction(transaction_params)

    # 對交易數據進行簽名
    signed_transaction = w3.eth.account.sign_transaction(transaction_data, receiver_private_key)

    # 發送交易
    transaction_hash = w3.eth.send_raw_transaction(signed_transaction.rawTransaction)

    # 等待交易被打包和確認
    transaction_receipt = w3.eth.wait_for_transaction_receipt(transaction_hash)

except ValueError as e:
    print(f"Transaction failed: {e}")

# 檢查交易是否成功
if transaction_receipt['status'] == 1:
    print("accepted 成功")
else:
    print("失敗")
    
balance(sender_address)
balance(receiver_address)
get_stage()
get_smartcontract_balance()
```

```python
# revealKey後得到key
key  = contract.functions.getKey().call()
print(key)

with open("key.txt", "wb") as f:
    f.write(key)
```

```python
# Step 4.1 (ok)，合約將錢給sender

# 獲取交易的nonce值
nonce = w3.eth.get_transaction_count(receiver_address)

# 構建交易參數
transaction_params = {
    'gas': 500000,
    'gasPrice': w3.to_wei('10', 'gwei'),
    'nonce': nonce,
    'from': receiver_address
}
try:
    # 構建交易數據
    transaction_data = contract.functions.noComplain().build_transaction(transaction_params)

    # 對交易數據進行簽名
    signed_transaction = w3.eth.account.sign_transaction(transaction_data, receiver_private_key)

    # 發送交易
    transaction_hash = w3.eth.send_raw_transaction(signed_transaction.rawTransaction)

    # 等待交易被打包和確認
    transaction_receipt = w3.eth.wait_for_transaction_receipt(transaction_hash)
    
except ValueError as e:
    print(f"Transaction failed: {e}")
    

# 檢查交易是否成功
if transaction_receipt['status'] == 1:
    print("noComplain 成功")
    print("交易成功")
else:
    print("交易失敗")
    
balance(sender_address)
balance(receiver_address)
get_stage()
get_smartcontract_balance()

#print('price:', contract.functions['price']().call(), "(Wei)")
```

```python
# 客戶生成的投訴資料
import json

with open('verify_item.json', 'r', encoding='utf-8') as f:
    loaded_verify_item = json.load(f)
#print(loaded_verify_item)
#print()

decrypted_data=[]
with open("decrypted_data.txt", "rb") as f:
    for line in f:
        decrypted_data.append(line.strip())
#print(decrypted_data)

indexIn = loaded_verify_item['indexIn']
indexOut = loaded_verify_item['indexOut']
root = w3.to_hex(w3.to_bytes(hexstr=loaded_verify_item['root']))
#proof = [w3.to_hex(w3.to_bytes(hexstr=item)) for item in loaded_verify_item['proofs']]
proofZin = [w3.to_hex(w3.to_bytes(hexstr=item)) for item in loaded_verify_item['proofZin']]
proofZout = [w3.to_hex(w3.to_bytes(hexstr=item)) for item in loaded_verify_item['proofZout']]

input = [decrypted_data[1], decrypted_data[2]]
output = decrypted_data[0]
#input = loaded_verify_item['inputsZ']
#output = loaded_verify_item['outputZ']

inputZH = [w3.to_hex(w3.to_bytes(hexstr=item)) for item in loaded_verify_item['inputZH']]
outputZH = w3.to_hex(w3.to_bytes(hexstr=loaded_verify_item['outputZH']))

print("indexIn:", indexIn)
print("\nindexOut:", indexOut)
print("\nroot:", root)
print("\nproofZin:", proofZin)
print("\nproofZout:", proofZout)
print("\ninput:", input)
print("\noutput:", output)
print("\ninputZH:", inputZH)
print("\noutputZH:", outputZH)

print(ciphertextRoot == root)
```

```python
# Step 4.2 (complain) 投訴檔案錯誤

# 獲取交易的nonce值
nonce = w3.eth.get_transaction_count(receiver_address)

# 構建交易參數
transaction_params = {
    'gas': 500000,
    'gasPrice': w3.to_wei('10', 'gwei'),
    'nonce': nonce,
    'from': receiver_address
}
try:
    # 構建交易數據 
    transaction_data = contract.functions.complain(indexIn, indexOut, root, proofZin, proofZout, input, output, inputZH, outputZH).build_transaction(transaction_params)
    #transaction_data = contract.functions.complain_request_ver(index, root, proof, input[0], input[1], output).build_transaction(transaction_params)

    # 對交易數據進行簽名
    signed_transaction = w3.eth.account.sign_transaction(transaction_data, receiver_private_key)

    # 發送交易
    transaction_hash = w3.eth.send_raw_transaction(signed_transaction.rawTransaction)

    # 等待交易被打包和確認
    transaction_receipt = w3.eth.wait_for_transaction_receipt(transaction_hash)
    
except ValueError as e:
    print(f"Transaction failed: {e}")
    
# complain 結果
result = contract.functions.getComplainResult().call()
print(result)
    
balance(sender_address)
balance(receiver_address)
get_stage()
get_smartcontract_balance()
```
