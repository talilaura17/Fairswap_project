```python
import math

def findIndex(m, y, s, index_m):
    n = len(y)
    
    if n == 1:
        return index_m
    
    index = y.index(s)
    
    if index <= (n - 1) // 2:
        index_m -= m / 2
        return findIndex(m / 2, y[1:(n - 1) // 2 + 1], s, index_m)
    else:
        return findIndex(m / 2, y[(n - 1) // 2 + 1:], s, index_m)

# Test findIndex()
a1 = ["y7", "y5", "y1", "y2", "y6", "y3", "y4"]
a2 = ["y1", "y2", "y3", "y4", "y5", "y6", "y7", "y8", "y9", "y10", "y11", "y12", "y13", "y14", "y15"]
a3 = ['y1', 'y2', 'y3', 'y4', 'y5', 'y6', 'y7', 'y8', 'y9', 'y10', 'y11', 'y12', 'y13', 'y14', 'y15', 'y16', 'y17', 'y18', 'y19', 'y20', 'y21', 'y22', 'y23', 'y24', 'y25', 'y26', 'y27', 'y28', 'y29', 'y30', 'y31']

# 葉節點數量m
m = math.pow(2, math.log2(len(a2) + 1) - 1)
print("Initial m:", m)

index_m = m
result_index_m = findIndex(m, a2, "y15", index_m)

print("Final index_m:", result_index_m-1)
```
